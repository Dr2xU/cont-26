# UMAP — Projet Final

## 1. Chargement des donnees

```python
import warnings
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from umap import UMAP

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')

RANDOM_STATE = 42
CANDIDATE_ROOTS = [Path.cwd(), Path.cwd().parent]
PROJECT_ROOT = next((p for p in CANDIDATE_ROOTS if (p / 'data' / 'city_lifestyle_dataset.csv').exists()), None)
if PROJECT_ROOT is None:
    raise FileNotFoundError("Impossible de localiser le dossier projet contenant data/city_lifestyle_dataset.csv")

DATA_PATH = PROJECT_ROOT / 'data' / 'city_lifestyle_dataset.csv'
OUTPUT_DIR = PROJECT_ROOT / 'outputs'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(DATA_PATH)
print(f'Shape: {df.shape}')
df.head()
```

## 2. Preprocessing

On sélectionne uniquement les colonnes numériques et on standardise les données (moyenne 0, écart-type 1), car UMAP est sensible aux échelles des variables.

```python
# Colonnes numériques (on exclut city_name et country)
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
X = df[numeric_cols].copy()

if X.isna().any().any():
    raise ValueError('Valeurs manquantes detectees dans les colonnes numeriques. Nettoyer avant UMAP.')

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f'Variables numeriques: {len(numeric_cols)}')
print(f'Shape de X: {X.shape}')
```

## 3. Application de UMAP et visualisation

```python
umap_model = UMAP(
    n_components=2,
    n_neighbors=15,
    min_dist=0.1,
    metric='euclidean',
    random_state=42
)

X_umap = umap_model.fit_transform(X_scaled)

# Visualisation colorée par continent/country
fig, ax = plt.subplots(figsize=(10, 7))

countries = df['country'].unique()
palette = sns.color_palette('tab10', n_colors=len(countries))

for i, country in enumerate(countries):
    mask = df['country'] == country
    ax.scatter(X_umap[mask, 0], X_umap[mask, 1],
               label=country, color=palette[i], alpha=0.7, s=40, edgecolors='w', linewidths=0.3)

ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')
ax.set_title('Projection UMAP (2D) — City Lifestyle Dataset')
ax.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.savefig('../outputs/umap_2d.png', dpi=150, bbox_inches='tight')
plt.show()


# Visualisation colorée par happiness_score pour voir une autre dimension
fig, ax = plt.subplots(figsize=(10, 7))

scatter = ax.scatter(X_umap[:, 0], X_umap[:, 1],
                     c=df['happiness_score'], cmap='viridis', alpha=0.7, s=40, edgecolors='w', linewidths=0.3)

plt.colorbar(scatter, label='Happiness Score')
ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')
ax.set_title('Projection UMAP — colorée par Happiness Score')
plt.tight_layout()
plt.savefig('../outputs/umap_happiness.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Observation

La projection UMAP révèle plusieurs groupes (clusters) de villes qui partagent des caractéristiques similaires en termes de qualité de vie.

- On observe des regroupements assez nets selon la variable `country`, ce qui suggère que les villes d'un même continent partagent des profils de vie similaires (densité, revenus, transports, etc.).
- La coloration par `happiness_score` montre un gradient au sein de certains clusters, et un gradient très claire en générale, de gauche (les plus bas) à droite (les plus hauts).

En résumé, UMAP fournit une visualisation cohérente et interprétable, adaptée pour identifier les profils de villes similaires dans ce dataset multidimensionnel.

## 5. Export des données 2D

```python
df_umap = pd.DataFrame(X_umap, columns=['UMAP_1', 'UMAP_2'])
df_umap['city_name'] = df['city_name'].values
df_umap['country'] = df['country'].values

df_umap.to_csv(f'{OUTPUT_DIR}/umap_2d.csv', index=False)
print("Export terminé : outputs/umap_2d.csv")
df_umap.head()
```
